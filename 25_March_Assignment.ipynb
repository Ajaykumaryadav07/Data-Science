```python
#Q1:
import pandas as pd

# 
flight_data=pd.read_excel('flight_price.xlsx')
#df.head()# Replace 'flight_prices.csv' with your file path if needed

# Get the dimensions of the dataset
rows, columns = flight_data.shape

# Print the dimensions
print("Number of rows:", rows)
print("Number of columns:", columns)
  
```

```python

import matplotlib.pyplot as plt

# Plot the histogram
plt.hist(flight_data['Price'], bins=20, edgecolor='black')

# Set the labels and title
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')

# Display the histogram
plt.show()
```

```python
flight_data.head()
```

```python
#Q#:
minimum_price = flight_data['Price'].min()
maximum_price = flight_data['Price'].max()

# Print the results
print("Minimum price:", minimum_price)
print("Maximum price:", maximum_price)
```

```python
#Q4
# Create a boxplot
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed
flight_data.boxplot(column='Price', by='Airline', vert=False)

# Set the labels and title
plt.xlabel('Price')
plt.ylabel('Airline')
plt.title('Price Variation by Airline')

# Rotate the y-axis labels if needed
plt.yticks(rotation=45)

# Display the boxplot
plt.show()
```

```python
#Q9:
playstore_df = pd.read_csv('https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv')

# Get the dimensions of the dataset
num_rows, num_cols = playstore_df.shape

# Print the dimensions
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
import matplotlib.pyplot as plt
```

```python
#Q10
# Load the dataset
#playstore_df = pd.read_csv('googleplaystore.csv')

# Create a boxplot of ratings by category
plt.figure(figsize=(10, 6))
playstore_df.boxplot(column='Rating', by='Category', rot=90)
plt.title('Rating of Apps by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
# Check for missing values
missing_values = playstore_df.isnull().sum()
# Print the missing values
print(missing_values)
```

```python
# Convert the 'Size' column to numeric values
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: float(x[:-1]) if x[-1] == 'M' else float(x[:-1])/1024)

# Create a scatter plot of app size vs rating
plt.figure(figsize=(10, 6))
plt.scatter(playstore_df['Size'], playstore_df['Rating'])
plt.title('App Size vs Rating')
plt.xlabel('Size (in MB)')
plt.ylabel('Rating')
plt.show()
```
