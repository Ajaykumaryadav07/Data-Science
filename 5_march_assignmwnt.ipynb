```python
#Q:1
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
x = [1,2,3,4,5]
y = [6,7,2,4,5]
```

```python
from bokeh.core.properties import Size
p = figure(title="Bokeh Circle Plot", x_axis_label="X", y_axis_label="Y")
p.circle(x,y, size=15, color="navy", alpha=0.5)
output_notebook()
show(p)
```

```python
#Q:2
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
source = ColumnDataSource(data=dict(x=x, y=y))
p = figure(title='Circle Glyph Examle')
p.circle('x', 'y', size=10,source=source)
output_file('circle.html')
show(p)
```

```python
p = figure(title='Example Plot', x_axis_label='X-axis label',
           y_axis_label='Y-axis label',
           x_axis_type='linear',
            y_axis_type='linear',
           x_range=(0,6),
           y_range=(0,8)
           )
p.circle('x', 'y', size=10,source=source,legend_label='Data points')
# customize the legend
p.legend.location = 'top_left'
p.legend.visible = True

# customize the title
p.title.text_color = 'blue'
p.title.align = 'center'

# customize the axes
p.xaxis.axis_label_text_color = 'green'
p.yaxis.axis_label_text_color = 'red'
p.xaxis.major_label_text_color = 'green'
p.yaxis.major_label_text_color = 'red'

# specify output file
output_file('example.html')

# show the plot
show(p)
```

```python
#Q:5
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # create the Bokeh plot
    p = figure()
    p.line([1,2,3], [4,5,6])

    # generate the HTML components for the plot
    script, div = components(p)

    # pass the components to the template
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run(debug=True)
from django.shortcuts import render
from bokeh.plotting import figure
from bokeh.embed import components

def index(request):
    # create the Bokeh plot
    p = figure()
    p.line([1,2,3], [4,5,6])

    # generate the HTML components for the plot
    script, div = components(p)

    # pass the components to the template
    return render(request, 'index.html', {'script': script, 'div': div})

```
