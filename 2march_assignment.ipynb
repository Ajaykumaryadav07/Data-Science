Matplotlib is a popular Python library used for creating 2D plots and graphs. It provides a variety of tools for creating static, animated, and interactive visualizations in Python.

Matplotlib is used for data visualization and exploration, scientific and engineering plots, financial and economic data visualization, and more.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot: It is used to plot a line connecting data points. It is useful for visualizing trends and patterns in data.

Scatter plot: It is used to plot individual data points on a two-dimensional plane. It is useful for visualizing the relationship between two variables.

Bar plot: It is used to plot categorical data with rectangular bars. It is useful for visualizing comparisons between different categories.

Histogram: It is used to plot the frequency distribution of a set of continuous or discrete data. It is useful for visualizing the distribution of data.

Pie chart: It is used to plot data as slices of a circle. It is useful for visualizing the proportion of different categories in a dataset.

```python
#Q:2
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#Q:3
import numpy as np
import matplotlib.pyplot as plt

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
fig, axs = plt.subplots(2, 2)

# Plot lines
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Add titles
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Add x and y labels
for ax in axs.flat:
    ax.set(xlabel='X', ylabel='Y')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
#Q:4
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Plotting a bar plot
plt.bar(company, profit)
plt.title("Company profits")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")
plt.show()

# Plotting a horizontal bar plot
plt.barh(company, profit)
plt.title("Company profits")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")
plt.show()
```

```python
#Q:5
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Plotting a box plot
plt.boxplot([box1, box2])
plt.title("Box plot")
plt.xlabel("Box")
plt.ylabel("Value")
plt.xticks([1, 2], ["Box 1", "Box 2"])
plt.show()
```
