```python
#Q:1
import scipy.stats as stats

sample_mean = 50
sample_std = 5
n = 100  # assuming sample size of 100

alpha = 0.05  # 95% confidence interval
z_critical = stats.norm.ppf(1 - alpha/2)

margin_of_error = z_critical * (sample_std / (n**0.5))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(confidence_interval)
```

```python
#Q:2
import numpy as np
import scipy.stats as stats

observed_frequencies = [16, 22, 18, 10, 12, 22]  # observed frequencies of blue, orange, green, yellow, red, brown
expected_frequencies = [0.2, 0.2, 0.2, 0.1, 0.1, 0.2]  # expected frequencies of blue, orange, green, yellow, red, brown
n = sum(observed_frequencies)

chi_square, p_value = stats.chisquare(observed_frequencies, np.array(expected_frequencies)*n)

alpha = 0.05

if p_value < alpha:
    print("Reject null hypothesis. The distribution of colors in the bag does not match the expected distribution.")
else:
    print("Fail to reject null hypothesis. The distribution of colors in the bag matches the expected distribution.")
```

```python
#Q:3
import numpy as np
from scipy.stats import chi2_contingency

# create contingency table
observed = np.array([[20, 15], [10, 25], [15, 20]])

# conduct chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(observed)

# print results
print("Chi-square statistic: {:.2f}".format(chi2_stat))
print("P-value: {:.4f}".format(p_val))
```

```python
#Q:5
import scipy.stats as stats

data = [75]  # sample data
n = len(data)
mean = sum(data) / n
std_dev = 12

alpha = 0.1  # 90% confidence level
t_value = stats.t.ppf(1 - alpha/2, df=n-1)

lower_bound = mean - t_value * std_dev / (n**0.5)
upper_bound = mean + t_value * std_dev / (n**0.5)

print("90% confidence interval: ({0:.4f}, {1:.4f})".format(lower_bound, upper_bound))
```

```python
#Q:6
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df = 10

x = np.linspace(0, 30, 500)  # set limits for the x-axis
y = chi2.pdf(x, df)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.fill_between(x[x >= 15], y[x >= 15], color='orange', alpha=0.5)
ax.set_xlim([0, 30])
ax.set_ylim([0, 0.25])
ax.set_xlabel('Chi-Square Statistic')
ax.set_ylabel('Probability Density')
ax.set_title('Chi-Square Distribution with 10 Degrees of Freedom')
plt.show()
```

To conduct a chi-square goodness of fit test, we need to first define our null and alternative hypotheses:

Null hypothesis (H0): The observed frequencies of heads and tails are not significantly different from the expected frequencies of a fair coin (i.e., p(heads) = 0.5 and p(tails) = 0.5).

Alternative hypothesis (Ha): The observed frequencies of heads and tails are significantly different from the expected frequencies of a fair coin.

We can use the chi-square goodness of fit test to determine if the observed frequencies are significantly different from the expected frequencies. The test statistic is calculated as follows:

χ2 = ∑((O - E)2 / E)

Where:
O = observed frequency
E = expected frequency

For a fair coin, we would expect to see 50 heads and 50 tails out of 100 flips. Therefore, the expected frequencies are:

E(heads) = 50
E(tails) = 50

The observed frequencies are:

O(heads) = 100 - 45 = 55
O(tails) = 45

Substituting these values into the formula, we get:

χ2 = ((55-50)2 / 50) + ((45-50)2 / 50)
= 1

Using a chi-square distribution table with 1 degree of freedom and a significance level of 0.05, we find that the critical value is 3.84.

Since our calculated χ2 value (1) is less than the critical value (3.84), we fail to reject the null hypothesis. Therefore, we do not have sufficient evidence to suggest that the observed frequencies are significantly different from the expected frequencies of a fair coin
