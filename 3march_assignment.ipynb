```python
#Q:2
import seaborn as  sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset("fmri")
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

```python
#Q:3
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
```

```python
fig, axes = plt.subplots(1,2, figsize=(12,6))
sns.boxplot(x='pclass', y='age',data=titanic,ax=axes[0])
sns.boxplot(x='pclass', y='age',data=titanic,ax=axes[1])
plt.show()
```

```python
diamond = sns.load_dataset('diamonds')
```

```python
sns.histplot(data=diamond, x='price', hue='cut',kde='True')
```

```python
#Q:5
iris = sns.load_dataset('iris')
sns.pairplot(iris,hue='species')
```

```python
#Q:6
flights=sns.load_dataset('flights')
```

```python
flights_matrix = flights.pivot('month', 'year', 'passengers')
sns.heatmap(flights_matrix, cmap='YlOrRd', annot=True, fmt='d')
```
